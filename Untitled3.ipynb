```python
import os
import cv2
import numpy as np
from PIL import Image

# إنشاء مجلدات البيانات
base_path = 'dataset'
os.makedirs(f'{base_path}/good', exist_ok=True)
os.makedirs(f'{base_path}/defect', exist_ok=True)

def create_synthetic_bottle(is_defective=False):
    # إنشاء صورة سوداء (خلفية)
    img = np.zeros((224, 224, 3), dtype=np.uint8)
    # رسم شكل زجاجة (أسطوانة بيضاء)
    cv2.ellipse(img, (112, 112), (50, 100), 0, 0, 360, (255, 255, 255), -1)

    if is_defective:
        # إضافة "عيب" (خط أسود كسر أو خدش)
        cv2.line(img, (90, 80), (130, 100), (0, 0, 0), 5)
        cv2.circle(img, (110, 150), 10, (0, 0, 0), -1) # بقعة تلوث

    return Image.fromarray(img)

# توليد 100 صورة سليمة و 100 صورة معيبة
print("جاري إنشاء صور الزجاجات برمجياً...")
for i in range(100):
    good_img = create_synthetic_bottle(is_defective=False)
    good_img.save(f'{base_path}/good/bottle_{i}.png')

    defect_img = create_synthetic_bottle(is_defective=True)
    defect_img.save(f'{base_path}/defect/defect_{i}.png')

print(f"✅ تم إنشاء {len(os.listdir(f'{base_path}/good'))} صورة سليمة")
print(f"✅ تم إنشاء {len(os.listdir(f'{base_path}/defect'))} صورة معيبة")
```

```python
import torch
import torch.nn as nn
from torchvision import models

# تحميل الموديل مدرب مسبقاً
model = models.resnet34(weights='DEFAULT')

# تعديل الطبقة الأخيرة (Classifier)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2) # مخرجان فقط (سليم أو معيب)

# نقل الموديل للمعالج الرسومي (GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print(f"✅ الموديل جاهز ويعمل على: {device}")
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

# تحويل الصور لمقاس وحجم موحد
data_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# قراءة المجلدات
full_dataset = datasets.ImageFolder('dataset', transform=data_transforms)

# التقسيم
train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_dataset, test_dataset = random_split(full_dataset, [train_size, test_size])

# إنشاء الـ Loaders (الحزام الناقل للبيانات)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

print(f"✅ تم تقسيم البيانات: {len(train_dataset)} تدريب، {len(test_dataset)} اختبار.")
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

model.train()
for epoch in range(10):
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/10 - Loss: {running_loss/len(train_loader):.4f}")

print("✅ انتهى التدريب!")
```

```python
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image
import matplotlib.pyplot as plt
import numpy as np

# وضع الموديل في نمط الاختبار
model.eval()

# 1. اختيار صورة معيبة من الاختبار
for i in range(len(test_dataset)):
    img_tensor, label = test_dataset[i]
    if label == 0: # 0 عادة ما يكون 'defect' حسب ترتيب الحروف
        break

input_batch = img_tensor.unsqueeze(0).to(device)

# 2. تطبيق Grad-CAM
target_layers = [model.layer4[-1]] # الطبقة الأخيرة
cam = GradCAM(model=model, target_layers=target_layers)
grayscale_cam = cam(input_tensor=input_batch, targets=[ClassifierOutputTarget(label)])[0, :]

# 3. عرض الصورة الأصلية مع الخريطة الحرارية
img_show = img_tensor.permute(1, 2, 0).cpu().numpy()
img_show = (img_show - img_show.min()) / (img_show.max() - img_show.min())
visualization = show_cam_on_image(img_show, grayscale_cam, use_rgb=True)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img_show)
plt.subplot(1, 2, 2)
plt.title("Grad-CAM Explanation")
plt.imshow(visualization)
plt.show()
```

```python
import gradio as gr
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image
import numpy as np
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image

# 1. تجهيز الموديل وتحميل الأوزان المحفوظة
def load_my_model():
    model = models.resnet34()
    model.fc = nn.Linear(model.fc.in_features, 2)
    # تحميل الأوزان (تأكد من وجود ملف final_model.pth في ملفات الكولاب)
    model.load_state_dict(torch.load('final_model.pth', map_location='cpu'))
    model.eval()
    return model

model_inference = load_my_model()

# 2. دالة المعالجة التي تعمل عند رفع الصورة
def predict_and_explain(input_img):
    # تحويل الصورة لتنسيق يناسب الموديل
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    img_pil = Image.fromarray(input_img)
    input_tensor = transform(img_pil).unsqueeze(0)

    # التنبؤ بالنتيجة
    with torch.no_grad():
        output = model_inference(input_tensor)
        _, pred = torch.max(output, 1)
        conf = torch.nn.functional.softmax(output, dim=1)[0]

    label = "معيبة (Defective) ❌" if pred.item() == 0 else "سليمة (Good) ✅"
    score = f"نسبة التأكد: {conf[pred.item()].item()*100:.2f}%"

    # تشغيل Grad-CAM لتوضيح سبب القرار
    target_layers = [model_inference.layer4[-1]]
    cam = GradCAM(model=model_inference, target_layers=target_layers)
    grayscale_cam = cam(input_tensor=input_tensor, targets=[ClassifierOutputTarget(pred.item())])[0, :]

    # دمج الخريطة الحرارية مع الصورة الأصلية
    img_np = np.array(img_pil.resize((224, 224))) / 255.0
    visualization = show_cam_on_image(img_np, grayscale_cam, use_rgb=True)

    return label, score, visualization

# 3. بناء واجهة الموقع (UI)
with gr.Blocks(title="AI Bottle Inspector") as demo:
    gr.Markdown("# 🔍 نظام فحص الزجاجات الذكي")
    gr.Markdown("ارفع صورة الزجاجة ليقوم الموديل بفحصها وشرح مكان العيب.")

    with gr.Row():
        with gr.Column():
            input_image = gr.Image(label="ارفع صورة هنا")
            btn = gr.Button("بدء الفحص")
        with gr.Column():
            result_label = gr.Textbox(label="النتيجة")
            confidence_label = gr.Textbox(label="الثقة")
            output_image = gr.Image(label="شرح القرار (Grad-CAM)")

    btn.click(fn=predict_and_explain, inputs=input_image, outputs=[result_label, confidence_label, output_image])

# 4. تشغيل الموقع مع رابط خارجي
demo.launch(share=True, debug=True)
```
